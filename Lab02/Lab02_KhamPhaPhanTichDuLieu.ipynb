Với các cột có kiểu dữ liệu số, bạn sẽ tính:
- Tỉ lệ % (từ 0 đến 100) các giá trị thiếu

- Giá trị median (phân vị 50)

- Giá trị max

Bạn sẽ lưu kết quả vào DataFrame `nume_col_info_df`, trong đó: 
- Tên của các cột là tên của các cột số trong `survey_df`
- Tên của các dòng là: "missing_percentage", "min", "lower_quartile", "median", "upper_quartile", "max"  

Để dễ nhìn, tất cả các giá trị bạn đều làm tròn với 1 chữ số thập phân bằng phương thức `.round(1)`.

````python
# YOUR CODE HERE
# raise NotImplementedError()

# Slice columns with numeric values
temp = survey_df.dtypes.apply(pd.api.types.is_numeric_dtype)
numeric_cols = temp.loc[temp == True].index
temp = survey_df[numeric_cols].copy()

def compute_stats(col):
    return [
        round(np.sum(col.isna()) / col.size * 100, 1),
        np.min(col),
        np.nanquantile(col, 0.25),
        np.nanmedian(col),    
        np.nanquantile(col, 0.75),
        np.max(col),
    ]
nume_col_info_df = temp.apply(compute_stats, axis='index')
nume_col_info_df.index = ['missing_percentage', 'min', 'lower_quartile', 'median', 'upper_quartile', 'max']
nume_col_info_df
````

````python
# YOUR CODE HERE
# raise NotImplementedError()
age = survey_df['Age']
yearscode = survey_df['YearsCode']
yearscodepro = survey_df['YearsCodePro']
workhours = survey_df['WorkWeekHrs']
max_hours_per_week = 7 * 24

num_invalid_Age_vals = age[(age < yearscode) | (age < yearscodepro)].size
num_invalid_WorkWeekHrs_vals = workhours[(workhours > max_hours_per_week)].size
num_invalid_YearsCode_vals = yearscode[(yearscode < yearscodepro) | (yearscode > age)].size
num_invalid_YearsCodePro_vals = yearscodepro[(yearscodepro > yearscode) | (yearscodepro > age)].size
````

````python
# YOUR CODE HERE
# raise NotImplementedError()

def is_invalid(row):
    age, yc, ycp, wh = row['Age'], row['YearsCode'], row['YearsCodePro'], row['WorkWeekHrs']
    b =  age < yc or age < ycp or wh > max_hours_per_week or yc < ycp
    return b

invalids = survey_df.apply(is_invalid, axis=1)
invalids = invalids[invalids]
survey_df.drop(labels=invalids.index, axis='index', inplace=True)
````

- Tỉ lệ % (từ 0 đến 100) của mỗi giá trị được sort theo tỉ lệ % giảm dần (ta không xét giá trị thiếu, tỉ lệ là tỉ lệ so với số lượng các giá trị không thiếu): bạn dùng dictionary để lưu, key là giá trị, value là tỉ lệ %; với cột mà ứng với câu hỏi dạng multichoice, cách làm tương tự như ở trên

Bạn sẽ lưu kết quả vào DataFrame `cate_col_info_df`, trong đó: 
- Tên của các cột là tên của các cột không phải số trong `survey_df`
- Tên của các dòng là: "missing_percentage", "num_values", "value_percentages"  

Để dễ nhìn, tất cả các giá trị bạn đều làm tròn với 1 chữ số thập phân bằng phương thức `.round(1)`.

Gợi ý: có thể bạn sẽ muốn dùng [phương thức `explode`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.explode.html).

````python
# Các cột ứng với câu hỏi khảo sát multichoice
multichoice_cols = ['DevType', 'Gender', 'JobFactors', 
                    'LanguageWorkedWith', 'LanguageDesireNextYear', 
                    'MiscTechWorkedWith', 'MiscTechDesireNextYear', 
                    'NEWCollabToolsWorkedWith', 'NEWCollabToolsDesireNextYear', 
                    'PlatformWorkedWith', 'PlatformDesireNextYear', 
                    'NEWStuck']
````

````python
pd.set_option('display.max_colwidth', 100) # Để dễ nhìn
pd.set_option('display.max_columns', None) # Để dễ nhìn
# YOUR CODE HERE
# raise NotImplementedError()


def compute_stats_categorical(col):
    missing_percentage = round(sum(col.isna()) / col.size * 100, 1)
    
    col = col.apply(func=lambda x : x.split(';') if type(x) == str else x)
    col = col.explode()
    
    value_counts = col.value_counts()
    num_values = value_counts.size
    value_dict = dict((value_counts / value_counts.sum() * 100).round(1))

    return [missing_percentage, num_values, value_dict]

q = survey_df[survey_df.columns.difference(numeric_cols)]
cate_col_info_df = q.apply(compute_stats_categorical, axis='index')
cate_col_info_df.index = ['missing_percentage', 'num_values', 'value_percentages']
cate_col_info_df
````

## Tiền xử lý để chuẩn bị dữ liệu cho bước phân tích để trả lời cho câu hỏi của bạn

Ta cắt `survey_df` thành bảng tạm `edlevel_jobsat_df` chỉ bao gồm những cột có liên quan. Theo đó, ta thực hiện các bước tiền xử lý sau:
* Rút ngắn các giá trị của `EdLevel` để thuận tiện
* Xóa đi các dòng có giá trị rỗng
* Định  nghĩa trước thứ tự cho `EdLevel` và `JobSat` (từ tệ nhất đến tốt nhất)

````python
# YOUR CODE HERE (OPTION)
shortened_edlevels = {
    'I never completed any formal education': 'No formal education',
    'Primary/elementary school': 'Primary school',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'Secondary school',
    'Some college/university study without earning a degree': 'College (no degree)',
    'Professional degree (JD, MD, etc.)': 'Professional degree',
    'Associate degree (A.A., A.S., etc.)': 'Associate degree',
    'Bachelor’s degree (B.A., B.S., B.Eng., etc.)': 'Bachelor\'s degree',
    'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)': 'Master\'s degree',
    'Other doctoral degree (Ph.D., Ed.D., etc.)': 'Doctoral degree',
}

edlevel_jobsat_df = survey_df[['EdLevel', 'JobSat']].copy().dropna()
edlevel_jobsat_df.EdLevel.replace(shortened_edlevels, inplace=True)

ordered_edlevels = [
    'No formal education',
    'Primary school',
    'Secondary school',
    'College (no degree)',
    'Professional degree',
    'Associate degree',
    'Bachelor\'s degree',
    'Master\'s degree',
    'Doctoral degree',
]

ordered_satlevels = [
    'Very dissatisfied',
    'Slightly dissatisfied',
    'Neither satisfied nor dissatisfied',
    'Slightly satisfied',
    'Very satisfied',
]
````

## Phân tích dữ liệu để ra câu trả lời cho câu hỏi của bạn (2đ)

Với mỗi trình độ học vấn, ta tính toán tỷ lệ phần trăm của 5 mức độ thỏa mãn (Very dissatisfied, ..., Very satisfied). Lưu kết quả này vào DataFrame `eljs_breakdown_df` (`eljs` là edlevel - jobsat):
* Đặt index là `EdLevel`
* Các cột ứng với từng mức độ thỏa mãn
* Index được sắp xếp theo `ordered_edlevels`
* Các cột được sắp xếp theo `ordered_satlevels`, từ trái sang phải

````python
# YOUR CODE HERE
# raise NotImplementedError()

def compute_satlevel_percentages(group):
    t = group.value_counts()
    t = t.reindex(ordered_satlevels)
    return list(t / len(group) * 100)

eljs_breakdown_df = edlevel_jobsat_df.groupby(by=['EdLevel']).aggregate(compute_satlevel_percentages)
eljs_breakdown_df = eljs_breakdown_df.reindex(ordered_edlevels)

temp = pd.DataFrame(list(eljs_breakdown_df.JobSat), index=eljs_breakdown_df.index, columns=ordered_satlevels)
eljs_breakdown_df = eljs_breakdown_df.join(temp)
eljs_breakdown_df.drop('JobSat', axis=1, inplace=True)
eljs_breakdown_df
````

Ta minh họa số liệu này bằng stacked bar chart.

````python
eljs_breakdown_df.plot(kind='bar', stacked=True, xlabel='Educational level', ylabel='Percentage', figsize=(10, 6))
plt.legend(loc='upper right')
````

Dựa trên số liệu này, ta nhận thấy:
* Mức độ thỏa mãn tăng theo trình độ học vấn
    * Nhóm người không học đầy đủ bất mãn nhất
    * Nhóm người có bằng Doctoral thỏa mãn nhất
* Hầu như không có sự khác biệt trong nhóm người có trình độ đại học dưới Doctoral (cụ thể là người có bằng Associate, Bachelor và Master)
* Trong số những người được học tập đàng hoàng, người có bằng Professional cảm thấy bất mãn nhất
