````python
# C√°c c·ªôt ·ª©ng v·ªõi c√¢u h·ªèi kh·∫£o s√°t multichoice
multichoice_cols = ['DevType', 'Gender', 'JobFactors', 
                    'LanguageWorkedWith', 'LanguageDesireNextYear', 
                    'MiscTechWorkedWith', 'MiscTechDesireNextYear', 
                    'NEWCollabToolsWorkedWith', 'NEWCollabToolsDesireNextYear', 
                    'PlatformWorkedWith', 'PlatformDesireNextYear', 
                    'NEWStuck']
````

Tr·∫£ l·ªùi ƒë∆∞·ª£c c√¢u h·ªèi n√†y gi√∫p ta bi·∫øt ƒë∆∞·ª£c r·∫±ng _c√≥ n√™n ƒë·∫ßu t∆∞ cho h·ªçc v·∫•n hay kh√¥ng_ n·∫øu nh∆∞ m·ª•c ti√™u c·ªßa ta l√† t√¨m ƒë∆∞·ª£c c√¥ng vi·ªác m√† m√¨nh c·∫£m th·∫•y y√™u th√≠ch v√† th·ªèa m√£n.

## Ti·ªÅn x·ª≠ l√Ω ƒë·ªÉ chu·∫©n b·ªã d·ªØ li·ªáu cho b∆∞·ªõc ph√¢n t√≠ch ƒë·ªÉ tr·∫£ l·ªùi cho c√¢u h·ªèi c·ªßa b·∫°n

Ta c·∫Øt `survey_df` th√†nh b·∫£ng t·∫°m `edlevel_jobsat_df` ch·ªâ bao g·ªìm nh·ªØng c·ªôt c√≥ li√™n quan. Theo ƒë√≥, ta th·ª±c hi·ªán c√°c b∆∞·ªõc ti·ªÅn x·ª≠ l√Ω sau:
* R√∫t ng·∫Øn c√°c gi√° tr·ªã c·ªßa `EdLevel` ƒë·ªÉ thu·∫≠n ti·ªán
* X√≥a ƒëi c√°c d√≤ng c√≥ gi√° tr·ªã r·ªóng
* ƒê·ªãnh  nghƒ©a tr∆∞·ªõc th·ª© t·ª± cho `EdLevel` v√† `JobSat` (t·ª´ t·ªá nh·∫•t ƒë·∫øn t·ªët nh·∫•t)

````python
# YOUR CODE HERE (OPTION)
shortened_edlevels = {
    'I never completed any formal education': 'No formal education',
    'Primary/elementary school': 'Primary school',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'Secondary school',
    'Some college/university study without earning a degree': 'College (no degree)',
    'Professional degree (JD, MD, etc.)': 'Professional degree',
    'Associate degree (A.A., A.S., etc.)': 'Associate degree',
    'Bachelor‚Äôs degree (B.A., B.S., B.Eng., etc.)': 'Bachelor\'s degree',
    'Master‚Äôs degree (M.A., M.S., M.Eng., MBA, etc.)': 'Master\'s degree',
    'Other doctoral degree (Ph.D., Ed.D., etc.)': 'Doctoral degree',
}

edlevel_jobsat_df = survey_df[['EdLevel', 'JobSat']].copy().dropna()
edlevel_jobsat_df.EdLevel.replace(shortened_edlevels, inplace=True)

ordered_edlevels = [
    'No formal education',
    'Primary school',
    'Secondary school',
    'College (no degree)',
    'Professional degree',
    'Associate degree',
    'Bachelor\'s degree',
    'Master\'s degree',
    'Doctoral degree',
]

ordered_satlevels = [
    'Very dissatisfied',
    'Slightly dissatisfied',
    'Neither satisfied nor dissatisfied',
    'Slightly satisfied',
    'Very satisfied',
]
````

---

## Ph√¢n t√≠ch d·ªØ li·ªáu ƒë·ªÉ ra c√¢u tr·∫£ l·ªùi cho c√¢u h·ªèi c·ªßa b·∫°n (2ƒë)

V·ªõi m·ªói tr√¨nh ƒë·ªô h·ªçc v·∫•n, ta t√≠nh to√°n t·ª∑ l·ªá ph·∫ßn trƒÉm c·ªßa 5 m·ª©c ƒë·ªô th·ªèa m√£n (Very dissatisfied, ..., Very satisfied). L∆∞u k·∫øt qu·∫£ n√†y v√†o DataFrame `eljs_breakdown_df` (`eljs` l√† edlevel - jobsat):
* ƒê·∫∑t index l√† `EdLevel`
* C√°c c·ªôt ·ª©ng v·ªõi t·ª´ng m·ª©c ƒë·ªô th·ªèa m√£n
* Index ƒë∆∞·ª£c s·∫Øp x·∫øp theo `ordered_edlevels`
* C√°c c·ªôt ƒë∆∞·ª£c s·∫Øp x·∫øp theo `ordered_satlevels`, t·ª´ tr√°i sang ph·∫£i

````python
# YOUR CODE HERE
# raise NotImplementedError()

def compute_satlevel_percentages(group):
    t = group.value_counts()
    t = t.reindex(ordered_satlevels)
    return list(t / len(group) * 100)

eljs_breakdown_df = edlevel_jobsat_df.groupby(by=['EdLevel']).aggregate(compute_satlevel_percentages)
eljs_breakdown_df = eljs_breakdown_df.reindex(ordered_edlevels)

temp = pd.DataFrame(list(eljs_breakdown_df.JobSat), index=eljs_breakdown_df.index, columns=ordered_satlevels)
eljs_breakdown_df = eljs_breakdown_df.join(temp)
eljs_breakdown_df.drop('JobSat', axis=1, inplace=True)
eljs_breakdown_df
````
